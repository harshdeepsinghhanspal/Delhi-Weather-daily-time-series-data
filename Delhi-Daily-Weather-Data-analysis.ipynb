```python
weather_data_train["date"] = pd.to_numeric(weather_data_train["date"].str.replace('-', ''))
```

```python
weather_data_train["date"].astype(float)
```

```python
print(weather_data_train)
```

```python
weather_data_train.drop(weather_data_train.index[weather_data_train['meanpressure'] >= 1200], inplace = True)
weather_data_train.drop(weather_data_train.index[weather_data_train['meanpressure'] <= 950], inplace = True)
```

```python
print(weather_data_train.describe())
```

```python
from sklearn import linear_model

X = weather_data_train[['humidity','wind_speed','meanpressure']]
y = weather_data_train['meantemp']

regr = linear_model.LinearRegression()
regr.fit(X, y)

predicted_meantemp = regr.predict([[50.0, 10, 1013]])

print(predicted_meantemp)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X, y)
model.score(X, y)
```

```python
model.coef_
```

```python
y_pred = model.predict(X)
```

```python
results = pd.DataFrame({'Actual': y, 'Predicted': y_pred})
print(results)
```

```python
plt.scatter(y, y_pred)
plt.xlabel('True Values ')
plt.ylabel('Predictions ')
plt.axis('equal')
plt.axis('square')
plt.xlim(xmin=0)
plt.ylim(ymin=0)
_ = plt.plot([-100, 100], [-100, 100])
```
