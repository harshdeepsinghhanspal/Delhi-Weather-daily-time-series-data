```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
weather_data_test = pd.read_csv('DailyDelhiClimateTest.csv', header=0, sep=",")
weather_data_train = pd.read_csv('DailyDelhiClimateTrain.csv', header=0, sep=",")
```

```python
print(weather_data_train)
```

```python
print(weather_data_train.describe())
```

```python
weather_data_train.plot(x ='date', y='meantemp', kind='line'),
plt.ylim(ymin=0)
plt.xlim(xmin=0)
plt.xticks(rotation=30)

plt.show()
```

```python
weather_data_train.plot(x ='date', y='humidity', kind='line',color="g"),
plt.ylim(ymin=0)
plt.xlim(xmin=0)
plt.xticks(rotation=30)

plt.show()
```

```python
weather_data_train.plot(x ='date', y='wind_speed', kind='line',color="orange"),
plt.ylim(ymin=0)
plt.xlim(xmin=0)
plt.xticks(rotation=30)

plt.show()
```

```python
weather_data_train.plot(x ='date', y='meanpressure', kind='line',color="red"),
plt.ylim(ymin=0)
plt.xlim(xmin=0)
plt.xticks(rotation=30)

plt.show()
```

```python
new_weather_data_train = weather_data_train[(weather_data_train['meanpressure'] <= 1200) & (weather_data_train['meanpressure'] >= 900)]

new_weather_data_train.plot(x ='date', y='meanpressure', kind='line',color="red"),
plt.ylim(ymin=0)
plt.xlim(xmin=0)
plt.xticks(rotation=30)

plt.show()
```

```python
new_weather_data_train = weather_data_train[(weather_data_train['meanpressure'] <= 1200) & (weather_data_train['meanpressure'] >= 900)]

new_weather_data_train.plot(x ='date', y='meanpressure', kind='line',color="red"),
plt.ylim(ymin=990)
plt.xlim(xmin=0)
plt.xticks(rotation=30)

plt.show()
```
